# Data Cleaning

Before we construct a knowledge graph from the data, as you can see, the data has multivalue in each column. We need to seperate them all and make one column consist of one value.

```python
df_tsel_summary_split = df_tsel['summary'].str.split(' at ', expand=True).iloc[:, :2]
df_tsel_summary_split.columns = ['current_position', 'current_company']

df_tsel_company_split = df_tsel['employments_employer_name'].str.split(',', expand=True).iloc[:, :10]
df_tsel_company_split.columns = [f'company_{i + 1}' for i in range(df_tsel_company_split.shape[1])]

df_tsel_title_split = df_tsel['employments_title'].str.split(',', expand=True).iloc[:, :10]
df_tsel_title_split.columns = [f'title_{i + 1}' for i in range(df_tsel_title_split.shape[1])]

df_tsel_skill_split = df_tsel['skills_name'].str.split(',', expand=True)
df_tsel_skill_split.columns = [f'skill_{i + 1}' for i in range(df_tsel_skill_split.shape[1])]

df_tsel_education_institution_split = df_tsel['educations_institution_name'].str.split(',', expand=True).iloc[:, :20]
df_tsel_education_institution_split = df_tsel_education_institution_split.apply(lambda row: pd.Series(row).drop_duplicates().tolist(), axis=1)
max_cols = df_tsel_education_institution_split.apply(lambda row: len(row)).max()
df_tsel_education_institution_split = pd.DataFrame(df_tsel_education_institution_split.tolist(), columns=[f'edu_institution_{i + 1}' for i in range(max_cols)])
df_tsel_education_institution_split = df_tsel_education_institution_split.dropna(axis=1, how='all')

df_tsel_education_major_split = df_tsel['educations_major_name'].str.split(',', expand=True)
df_tsel_education_major_split.columns = [f'edu_major_{i + 1}' for i in range(df_tsel_education_major_split.shape[1])]
df_tsel_education_major_split = df_tsel_education_major_split.applymap(lambda x: None if pd.notna(x) and any(char.isdigit() for char in str(x)) else x)

df_tsel_location_address_split = df_tsel['locations_address'].str.split(',', expand=True).iloc[:, :1]
df_tsel_location_address_split.columns = ['current_location_by_address']

df_tsel_location_city_split = df_tsel['locations_city_name'].str.split(',', expand=True).iloc[:, :1]
df_tsel_location_city_split.columns = ['current_location_by_city']

df_tsel_location_region_split = df_tsel['locations_region_name'].str.split(',', expand=True).iloc[:, :1]
df_tsel_location_region_split.columns = ['current_location_by_region']

df_tsel_location_country_split = df_tsel['locations_country_name'].str.split(',', expand=True).iloc[:, :1]
df_tsel_location_country_split.columns = ['current_location_by_country']

df_tsel_location_postalCode_split = df_tsel['locations_postalCode'].str.split(',', expand=True).iloc[:, :1]
df_tsel_location_postalCode_split.columns = ['current_location_by_postal_code']

df_tsel = pd.concat([df_tsel, df_tsel_summary_split, df_tsel_company_split, df_tsel_title_split,
                     df_tsel_skill_split, df_tsel_education_institution_split, df_tsel_education_major_split,
                     df_tsel_location_address_split, df_tsel_location_city_split, df_tsel_location_region_split,
                     df_tsel_location_country_split, df_tsel_location_postalCode_split], axis=1)

df_tsel = df_tsel.drop(['summary', 'employments_employer_name', 'employments_title',
                        'skills_name', 'educations_institution_name', 'educations_major_name',
                        'locations_address', 'locations_city_name', 'locations_region_name',
                        'locations_country_name', 'locations_postalCode'], axis=1)

df_tsel = df_tsel.fillna('unknown')

df_tsel = df_tsel.applymap(lambda x: str(x).replace(' ', '_'))\

df_tsel = df_tsel.replace({'unknown': None})
```
